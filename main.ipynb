```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Dataset.csv')
```

```python
df.head()
```

# Checking For Null Values

```python
df.isnull().sum()
```

# Checking For Coorelation

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(),linewidths=1,annot=True,cmap="coolwarm")
```

```python
# spliting the dependent & Independent Variables
from sklearn.preprocessing import StandardScaler
X=df.iloc[:,:-1].values
X=StandardScaler().fit_transform(X)
y=df.iloc[:,-1].values
```

```python
# features & target variable
print("Feature Data Dimensions:",X.shape)
print("Target Data Dimensions:",y.shape)
```

```python
# spliting the testing and training data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lreg = LogisticRegression()
lreg.fit(X_train, y_train)
y_pred = lreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy of the Logistic Regression Model: ",accuracy)
```

# Hyper Parameter Tunning

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV
model = LogisticRegression()
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l1','l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
# define grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(X,y)
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
Hyper=LogisticRegression(C=100,penalty='l2',solver='newton-cg')
Hyper.fit(X_train,y_train)
y_pred_new=Hyper.predict(X_test)
```

```python
print("Accuracy after tunning the Model: ",accuracy_score(y_test,y_pred_new)*100)
```
